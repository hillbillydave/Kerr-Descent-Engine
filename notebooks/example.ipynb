# Kerr-Descent-Engine Example Notebook

This notebook demonstrates how to run the Kerr Descent Engine, generate a trajectory, compute diagnostics, and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
from src import KerrDescentEngine, descent_step, compute_metrics
```

## Initialize the Engine

```python
engine = KerrDescentEngine(M=1.0, a=0.5, step_size=0.01)
initial_state = np.array([10.0, 1.0, 0.0])
```

## Run the Descent

```python
trajectory = engine.run(initial_state, steps=1000, descent_fn=descent_step)
trajectory[:5]
```

## Plot the Radial Descent

```python
r_values = trajectory[:, 0]
plt.figure(figsize=(10, 5))
plt.plot(r_values)
plt.title("Radial Descent into Kerr Interior")
plt.xlabel("Step")
plt.ylabel("r coordinate")
plt.grid(True)
plt.show()
```

## Compute Diagnostics for the Last Step

```python
metrics = compute_metrics(trajectory[-1], trajectory[-2])
metrics
```
